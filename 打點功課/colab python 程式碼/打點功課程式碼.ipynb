```python
 import json
import requests

def get_place_details(api_key, address, latitude, longitude):
    search_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    search_params = {
        "input": address,
        "inputtype": "textquery",
        "fields": "place_id,types",
        "locationbias": f"point:{latitude},{longitude}",
        "key": api_key
    }

    response = requests.get(search_url, params=search_params)
    results = response.json().get('candidates', [])

    if results:
        place_id = results[0].get('place_id', 'Not found')
        place_type = results[0].get('types', ['None'])[0]
    else:
        place_id = 'Not found'
        place_type = 'None'

    return place_id, place_type

def update_json_with_place_details(api_key, data):
    for entry in data:
        address = entry['fullAddress']
        latitude = entry['latitude']
        longitude = entry['longitude']

        place_id, place_type = get_place_details(api_key, address, latitude, longitude)

        entry['googleID'] = place_id
        entry['placeType'] = place_type

    return data

# Example usage
api_key = ''  # Replace with your actual Google API key

json_data = [
     {
        "affiliateId": "8a818cf519b813f80119b8141780002b",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "花蓮本會",
        "affiliateEngName": "",
        "buildingName": "靜思精舍",
        "affiliateAttribute": "本會",
        "postalCodeName": "花蓮縣新城鄉",
        "address": "康樂村精舍街88巷1號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8266779",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 971,
        "fullAddress": "971花蓮縣新城鄉康樂村精舍街88巷1號",
        "continents": "本會",
        "latitude": 24.0545366,
        "longitude": 121.6017492
    },
    {
        "affiliateId": "8a81958b2c7a010a012c7c7101bc0002",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "玉里聯絡處",
        "affiliateEngName": "",
        "buildingName": "玉里靜思堂",
        "affiliateAttribute": "聯絡處",
        "postalCodeName": "花蓮縣玉里鎮",
        "address": "民權街56號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8888816",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 981,
        "fullAddress": "981花蓮縣玉里鎮民權街56號",
        "continents": "本會",
        "latitude": 23.3347041,
        "longitude": 121.3158719
    },
    {
        "affiliateId": "8a81959866a08a36016a00cf89020007",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "光復共修處",
        "affiliateEngName": "",
        "buildingName": "光復會所",
        "affiliateAttribute": "共修處",
        "postalCodeName": "花蓮縣光復鄉",
        "address": "大華村中山路二段341號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2019/04/09",
        "modifyOpId": "朱李育",
        "province": "花蓮縣",
        "postalCode": 976,
        "fullAddress": "976花蓮縣光復鄉大華村中山路二段341號",
        "continents": "本會",
        "latitude": 23.6688767,
        "longitude": 121.4250737
    },
    {
        "affiliateId": "8a81958b2c7a010a012c7c7182da0003",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "吉安共修處",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "共修處",
        "postalCodeName": "花蓮縣吉安鄉",
        "address": "明仁三街305巷12號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2020/08/14",
        "stopOpId": "鄭鳳嘉",
        "modifyDate": "2020/08/14",
        "modifyOpId": "鄭鳳嘉",
        "province": "花蓮縣",
        "postalCode": 973,
        "fullAddress": "973花蓮縣吉安鄉明仁三街305巷12號",
        "continents": "本會",
        "latitude": 23.9871651,
        "longitude": 121.564985
    },
    {
        "affiliateId": "8a818cf519b813f80119b81417800036",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "志業中心",
        "affiliateEngName": "",
        "buildingName": "花蓮靜思堂",
        "affiliateAttribute": "分會",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路3段703號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "0912-412600 ext. 4002",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路3段703號",
        "continents": "本會",
        "latitude": 23.994869,
        "longitude": 121.590999
    },
    {
        "affiliateId": "8a81959866a08a3601699370d0470005",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "美崙聯絡點",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "聯絡點",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "民運里001鄰中美路242號 一樓",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2019/03/19",
        "modifyOpId": "王淑芬",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市民運里001鄰中美路242號 一樓",
        "continents": "本會",
        "latitude": 23.9948497,
        "longitude": 121.6290042
    },
    {
        "affiliateId": "8a8195985ec5c0bc0161d4c8f1f6005f",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "壽豐聯絡點",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "聯絡點",
        "postalCodeName": "花蓮縣壽豐鄉",
        "address": "壽豐路13號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2019/09/22",
        "stopOpId": "DBA",
        "modifyDate": "2018/02/27",
        "modifyOpId": "田智賢",
        "province": "花蓮縣",
        "postalCode": 974,
        "fullAddress": "974花蓮縣壽豐鄉壽豐路13號",
        "continents": "本會",
        "latitude": 23.8616569,
        "longitude": 121.5403927
    },
    {
        "affiliateId": "8a81958b3d9e177d013da55c53670002",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "鳳林共修處",
        "affiliateEngName": "",
        "buildingName": "鳳林靜思堂",
        "affiliateAttribute": "共修處",
        "postalCodeName": "花蓮縣鳳林鎮",
        "address": "信義路321號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8762578",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 975,
        "fullAddress": "975花蓮縣鳳林鎮信義路321號",
        "continents": "本會",
        "latitude": 23.7496566,
        "longitude": 121.453453
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00ee",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "玉里慈濟醫院",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "醫院",
        "postalCodeName": "花蓮縣玉里鎮",
        "address": "民權街1之1號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8882718",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 981,
        "fullAddress": "981花蓮縣玉里鎮民權街1之1號",
        "continents": "本會",
        "latitude": 23.3352975,
        "longitude": 121.3181733
    },
    {
        "affiliateId": "2D30382E35352E30352E353039303539",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "玉里慈濟居家式長照機構",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "長照機構",
        "postalCodeName": "花蓮縣玉里鎮",
        "address": "民權街1之1號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 981,
        "fullAddress": "981花蓮縣玉里鎮民權街1之1號",
        "continents": "本會",
        "latitude": 23.3352975,
        "longitude": 121.3181733
    },
    {
        "affiliateId": "8a81959866a08a3601693770108c0004",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "志學大愛農場",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "大愛農場",
        "postalCodeName": "花蓮縣壽豐鄉",
        "address": "志學村路內段400一401號",
        "canPublishExternal": "否",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/08/17",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 974,
        "fullAddress": "974花蓮縣壽豐鄉志學村路內段400一401號",
        "continents": "本會",
        "latitude": 23.9078055,
        "longitude": 121.5275518
    },
    {
        "affiliateId": "8a81958b312f09b101313113d6ae0022",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "花蓮慈院靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段707號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/07/16",
        "modifyOpId": "李青海",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段707號",
        "continents": "本會",
        "latitude": 23.9953743,
        "longitude": 121.5927399
    },
    {
        "affiliateId": "8a81958b312f09b1013131142c1b0023",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "花蓮靜思堂靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段703號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/07/16",
        "modifyOpId": "李青海",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段703號",
        "continents": "本會",
        "latitude": 23.994869,
        "longitude": 121.590999
    },
    {
        "affiliateId": "8a81958b312f09b1013130958e61000d",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "花蓮新城靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "花蓮縣新城鄉",
        "address": "精舍街71號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/07/16",
        "modifyOpId": "李青海",
        "province": "花蓮縣",
        "postalCode": 971,
        "fullAddress": "971花蓮縣新城鄉精舍街71號",
        "continents": "本會",
        "latitude": 24.0555358,
        "longitude": 121.5966344
    },
    {
        "affiliateId": "8a81958b312f09b10131310f13d9001b",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "精舍靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "花蓮縣新城鄉",
        "address": "康樂村精舍街88巷3號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/07/16",
        "modifyOpId": "李青海",
        "province": "花蓮縣",
        "postalCode": 971,
        "fullAddress": "971花蓮縣新城鄉康樂村精舍街88巷3號",
        "continents": "本會",
        "latitude": 24.0563125,
        "longitude": 121.6029644
    },
    {
        "affiliateId": "2D31312E30302E35382E353430333335",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "玉里靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "花蓮縣玉里鎮",
        "address": "民權街1-1號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 981,
        "fullAddress": "981花蓮縣玉里鎮民權街1-1號",
        "continents": "本會",
        "latitude": 23.3346713,
        "longitude": 121.3184139
    },
    {
        "affiliateId": "2D31312E31312E31322E363430343238",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "2D31312E31312E31322E363430343838",
        "affiliateName": "花蓮大愛幼兒園",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "幼兒園",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中山路一段3巷182號1樓之7",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2022/09/05",
        "modifyOpId": "user2",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中山路一段3巷182號1樓之7",
        "continents": "本會",
        "latitude": 23.9960891,
        "longitude": 121.5888481
    },
    {
        "affiliateId": "8a81958b2c7f7944012c8096f6c90044",
        "nation": "臺灣",
        "mission": "教育",
        "parentAffiliate": "2D31312E31312E31322E363430343832",
        "affiliateName": "慈濟科技大學",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "大學",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "建國路二段880號",
        "canPublishExternal": "是",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8572158",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市建國路二段880號",
        "continents": "本會",
        "latitude": 23.9967593,
        "longitude": 121.5647535
    },
    {
        "affiliateId": "8a81958b2c7f7944012c809736da0045",
        "nation": "臺灣",
        "mission": "教育",
        "parentAffiliate": "2D31312E31312E31322E363430343832",
        "affiliateName": "慈濟大學",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "大學",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段701號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8565301 ext. 1018",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段701號",
        "continents": "本會",
        "latitude": 23.9929379,
        "longitude": 121.5912799
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00eb",
        "nation": "臺灣",
        "mission": "教育",
        "parentAffiliate": "2D31312E31312E31322E363430343832",
        "affiliateName": "慈大附中(中學部)",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "中學",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "介仁街178號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8572823",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市介仁街178號",
        "continents": "本會",
        "latitude": 24.0009473,
        "longitude": 121.5754747
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00ec",
        "nation": "臺灣",
        "mission": "教育",
        "parentAffiliate": "2D31312E31312E31322E363430343832",
        "affiliateName": "慈大附小(小學部)",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "小學",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "介仁街178號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8578819",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市介仁街178號",
        "continents": "本會",
        "latitude": 24.0009473,
        "longitude": 121.5754747
    },
    {
        "affiliateId": "2D31312E31312E31322E363430343639",
        "nation": "臺灣",
        "mission": "教育",
        "parentAffiliate": "2D31312E31312E31322E363430343832",
        "affiliateName": "慈大附設幼兒園",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "幼兒園",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "介仁街178號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市介仁街178號",
        "continents": "本會",
        "latitude": 24.0009473,
        "longitude": 121.5754747
    },
    {
        "affiliateId": "2D30382E35352E30352E353039303632",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "玉里慈濟居家護理所",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "護理之家",
        "postalCodeName": "花蓮縣玉里鎮",
        "address": "民權街1之1號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 981,
        "fullAddress": "981花蓮縣玉里鎮民權街1之1號",
        "continents": "本會",
        "latitude": 23.3352975,
        "longitude": 121.3181733
    },
    {
        "affiliateId": "8a81958b320f654301321eceb59a0066",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "鳳林共修處環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "花蓮縣鳳林鎮",
        "address": "花蓮縣鳳林鎮信義路321號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2020/04/14",
        "modifyOpId": "張涵鈞",
        "province": "花蓮縣",
        "postalCode": 975,
        "fullAddress": "975花蓮縣鳳林鎮花蓮縣鳳林鎮信義路321號",
        "continents": "本會",
        "latitude": 23.7496566,
        "longitude": 121.453453
    },
    {
        "affiliateId": "8a81958b320f654301321ece51ea0060",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "玉里環保教育站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保教育站",
        "postalCodeName": "花蓮縣玉里鎮",
        "address": "花蓮縣玉里鎮忠孝路120號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2019/08/21",
        "modifyOpId": "張涵鈞",
        "province": "花蓮縣",
        "postalCode": 981,
        "fullAddress": "981花蓮縣玉里鎮花蓮縣玉里鎮忠孝路120號",
        "continents": "本會",
        "latitude": 23.3254506,
        "longitude": 121.3183646
    },
    {
        "affiliateId": "8a81958b320f654301321ece64ea0061",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "吉安共修處環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "花蓮縣吉安鄉",
        "address": "明仁三街305巷12號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2020/08/14",
        "stopOpId": "鄭鳳嘉",
        "modifyDate": "2020/08/14",
        "modifyOpId": "鄭鳳嘉",
        "province": "花蓮縣",
        "postalCode": 973,
        "fullAddress": "973花蓮縣吉安鄉明仁三街305巷12號",
        "continents": "本會",
        "latitude": 23.9871651,
        "longitude": 121.564985
    },
    {
        "affiliateId": "8a81958b320f654301321ece74620062",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "花蓮環保教育站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保教育站",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段381號",
        "canPublishExternal": "否",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2022/09/27",
        "modifyOpId": "si_j21436",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段381號",
        "continents": "本會",
        "latitude": 23.9830036,
        "longitude": 121.5853328
    },
    {
        "affiliateId": "8a81958b320f654301321ece98840064",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "瑞穗環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "花蓮縣瑞穗鄉",
        "address": "外環道與九如路交叉路旁",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2022/04/18",
        "stopOpId": "DBA",
        "modifyDate": "2014/03/17",
        "modifyOpId": "張涵鈞",
        "province": "花蓮縣",
        "postalCode": 978,
        "fullAddress": "978花蓮縣瑞穗鄉外環道與九如路交叉路旁",
        "continents": "本會",
        "latitude": 23.5205619,
        "longitude": 121.4113308
    },
    {
        "affiliateId": "8a81958b320f654301321ecea7780065",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b8141780002b",
        "affiliateName": "精舍環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "花蓮縣新城鄉",
        "address": "康樂村精舍街88巷1號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/08/31",
        "modifyOpId": "蔡奇珊",
        "province": "花蓮縣",
        "postalCode": 971,
        "fullAddress": "971花蓮縣新城鄉康樂村精舍街88巷1號",
        "continents": "本會",
        "latitude": 24.0545366,
        "longitude": 121.6017492
    },
    {
        "affiliateId": "8a81958b2c7f7944012c8096a4830043",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "8a81958b2c7f7944012c80888ba80009",
        "affiliateName": "慈濟骨髓幹細胞中心",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "骨髓幹細胞中心",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段701號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "蔡奇珊",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段701號",
        "continents": "本會",
        "latitude": 23.9929379,
        "longitude": 121.5912799
    },
    {
        "affiliateId": "8a81958b2c7f7944012c80888ba80009",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "花蓮慈濟醫院",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "醫學中心",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段707號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "03-8561825",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段707號",
        "continents": "本會",
        "latitude": 23.9953743,
        "longitude": 121.5927399
    },
    {
        "affiliateId": "2D30382E35352E30352E353039303131",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "花蓮慈濟居家護理所",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "護理之家",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段707號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段707號",
        "continents": "本會",
        "latitude": 23.9953743,
        "longitude": 121.5927399
    },
    {
        "affiliateId": "2D30382E35352E30352E353039303133",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "花蓮慈濟社區式長照機構",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "長照機構",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中山路一段3巷25弄36號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中山路一段3巷25弄36號",
        "continents": "本會",
        "latitude": 23.9968642,
        "longitude": 121.590534
    },
    {
        "affiliateId": "2D30382E35352E30352E353039303135",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "花蓮慈濟居家式長照機構",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "長照機構",
        "postalCodeName": "花蓮縣花蓮市",
        "address": "中央路三段707號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "花蓮縣",
        "postalCode": 970,
        "fullAddress": "970花蓮縣花蓮市中央路三段707號",
        "continents": "本會",
        "latitude": 23.9953743,
        "longitude": 121.5927399
    },
    {
        "affiliateId": "8a81958b47e53da60147fd37a0e20003",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "竹山聯絡處",
        "affiliateEngName": "",
        "buildingName": "竹山靜思堂",
        "affiliateAttribute": "聯絡處",
        "postalCodeName": "南投縣竹山鎮",
        "address": "延和里大智路52號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "049-2659671",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "南投縣",
        "postalCode": 557,
        "fullAddress": "557南投縣竹山鎮延和里大智路52號",
        "continents": "本會",
        "latitude": 23.7602475,
        "longitude": 120.6940623
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00bd",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "南投聯絡處",
        "affiliateEngName": "",
        "buildingName": "南投會所",
        "affiliateAttribute": "聯絡處",
        "postalCodeName": "南投縣南投市",
        "address": "果稟路101號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "049-2241169",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "南投縣",
        "postalCode": 540,
        "fullAddress": "540南投縣南投市果稟路101號",
        "continents": "本會",
        "latitude": 23.9059318,
        "longitude": 120.6932728
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00aa",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "埔里聯絡處",
        "affiliateEngName": "",
        "buildingName": "埔里會所",
        "affiliateAttribute": "聯絡處",
        "postalCodeName": "南投縣埔里鎮",
        "address": "西門里中正一路100號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "049-2997296",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "南投縣",
        "postalCode": 545,
        "fullAddress": "545南投縣埔里鎮西門里中正一路100號",
        "continents": "本會",
        "latitude": 23.9702423,
        "longitude": 120.9656994
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00ab",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "草屯聯絡處",
        "affiliateEngName": "",
        "buildingName": "草屯會所",
        "affiliateAttribute": "聯絡處",
        "postalCodeName": "南投縣草屯鎮",
        "address": "碧山路859號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "049-2323249",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "南投縣",
        "postalCode": 542,
        "fullAddress": "542南投縣草屯鎮碧山路859號",
        "continents": "本會",
        "latitude": 23.9754219,
        "longitude": 120.6651441
    },
    {
        "affiliateId": "8a81958b312f09b1013130966ac3000e",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "南投靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "南投縣南投市",
        "address": "振興里8鄰菓稟路101號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/07/16",
        "modifyOpId": "李青海",
        "province": "南投縣",
        "postalCode": 540,
        "fullAddress": "540南投縣南投市振興里8鄰菓稟路101號",
        "continents": "本會",
        "latitude": 23.9059318,
        "longitude": 120.6932728
    },
    {
        "affiliateId": "2D31312E30302E35382E353430323036",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "埔里聯絡處靜思小築",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思小築",
        "postalCodeName": "南投縣埔里鎮",
        "address": "西門里中正一路100號",
        "canPublishExternal": "否",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "南投縣",
        "postalCode": 545,
        "fullAddress": "545南投縣埔里鎮西門里中正一路100號",
        "continents": "本會",
        "latitude": 23.9702423,
        "longitude": 120.9656994
    },
    {
        "affiliateId": "2D31312E30302E35382E353430323536",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "草屯聯絡處靜思小築",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思小築",
        "postalCodeName": "南投縣草屯鎮",
        "address": "碧山路859號",
        "canPublishExternal": "否",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "南投縣",
        "postalCode": 542,
        "fullAddress": "542南投縣草屯鎮碧山路859號",
        "continents": "本會",
        "latitude": 23.9754219,
        "longitude": 120.6651441
    },
    {
        "affiliateId": "2D31312E30302E35382E353430333536",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "竹山聯絡處靜思小築",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思小築",
        "postalCodeName": "南投縣竹山鎮",
        "address": "延和里大智路52號",
        "canPublishExternal": "否",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "南投縣",
        "postalCode": 557,
        "fullAddress": "557南投縣竹山鎮延和里大智路52號",
        "continents": "本會",
        "latitude": 23.7602475,
        "longitude": 120.6940623
    },
    {
        "affiliateId": "8a81958b320f654301321eb7ec51000b",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "水里環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣水里鄉",
        "address": "永豐二路26號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2022/04/18",
        "stopOpId": "DBA",
        "modifyDate": "2015/07/31",
        "modifyOpId": "張涵鈞",
        "province": "南投縣",
        "postalCode": 553,
        "fullAddress": "553南投縣水里鄉永豐二路26號",
        "continents": "本會",
        "latitude": 23.8054817,
        "longitude": 120.856744
    },
    {
        "affiliateId": "8a81958b320f654301321eb80c11000c",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "北山環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣國姓鄉",
        "address": "北山坑段137-5.137-7地號等兩筆高架橋下",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/08/31",
        "modifyOpId": "蔡奇珊",
        "province": "南投縣",
        "postalCode": 544,
        "fullAddress": "544南投縣國姓鄉北山坑段137-5.137-7地號等兩筆高架橋下",
        "continents": "本會",
        "latitude": 23.98083,
        "longitude": 120.871201
    },
    {
        "affiliateId": "8a81958b320f654301321eb96c5d0013",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "竹山環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣竹山鎮",
        "address": "下崁段柯子坑小段122號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2022/04/18",
        "stopOpId": "DBA",
        "modifyDate": "2014/05/05",
        "modifyOpId": "蔡奇珊",
        "province": "南投縣",
        "postalCode": 557,
        "fullAddress": "557南投縣竹山鎮下崁段柯子坑小段122號",
        "continents": "本會",
        "latitude": 23.712201,
        "longitude": 120.6890055
    },
    {
        "affiliateId": "8a81958b320f654301321eba0517001b",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "南投聯絡處環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣南投市",
        "address": "稟路101號（南投園區內）",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2020/04/14",
        "modifyOpId": "張涵鈞",
        "province": "南投縣",
        "postalCode": 540,
        "fullAddress": "540南投縣南投市稟路101號（南投園區內）",
        "continents": "本會",
        "latitude": 23.93701,
        "longitude": 120.7007221
    },
    {
        "affiliateId": "8a8195985ec5c0bc0161ff8180a80062",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "南埔環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣草屯鎮",
        "address": "南投縣草屯鎮南埔里中正路381-20巷72號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2018/03/07",
        "modifyOpId": "田智賢",
        "province": "南投縣",
        "postalCode": 542,
        "fullAddress": "542南投縣草屯鎮南投縣草屯鎮南埔里中正路381-20巷72號",
        "continents": "本會",
        "latitude": 23.9774001,
        "longitude": 120.7189859
    },
    {
        "affiliateId": "8a81958b320f654301321eba61040020",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "草屯聯絡處環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣草屯鎮",
        "address": "碧山路859號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2020/04/14",
        "modifyOpId": "張涵鈞",
        "province": "南投縣",
        "postalCode": 542,
        "fullAddress": "542南投縣草屯鎮碧山路859號",
        "continents": "本會",
        "latitude": 23.9754219,
        "longitude": 120.6651441
    },
    {
        "affiliateId": "8a81958b320f654301321ebad40d0023",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "鹿谷環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣鹿谷鄉",
        "address": "中正路二段303巷1號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2022/04/18",
        "stopOpId": "DBA",
        "modifyDate": "2011/08/31",
        "modifyOpId": "蔡奇珊",
        "province": "南投縣",
        "postalCode": 558,
        "fullAddress": "558南投縣鹿谷鄉中正路二段303巷1號",
        "continents": "本會",
        "latitude": 23.7501482,
        "longitude": 120.74843
    },
    {
        "affiliateId": "8a81958b320f654301321ebaf6b80025",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "新庄環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣草屯鎮",
        "address": "新庄里芬草路",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "停用",
        "stopDate": "2022/04/18",
        "stopOpId": "DBA",
        "modifyDate": "2011/08/31",
        "modifyOpId": "蔡奇珊",
        "province": "南投縣",
        "postalCode": 542,
        "fullAddress": "542南投縣草屯鎮新庄里芬草路",
        "continents": "本會",
        "latitude": 23.9875551,
        "longitude": 120.6784077
    },
    {
        "affiliateId": "8a81958b320f654301321ebb20180027",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "藍城環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "南投縣埔里鎮",
        "address": "房里里房里路29-10號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2011/08/31",
        "modifyOpId": "蔡奇珊",
        "province": "南投縣",
        "postalCode": 545,
        "fullAddress": "545南投縣埔里鎮房里里房里路29-10號",
        "continents": "本會",
        "latitude": 23.9799545,
        "longitude": 120.9517331
    },
    {
        "affiliateId": "2D31312E34362E30322E363439383539",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a81958b3f824ef1013f8f43f2d00001",
        "affiliateName": "竹山聯絡處環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保教育站",
        "postalCodeName": "南投縣竹山鎮",
        "address": "大智路52號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "南投縣",
        "postalCode": 557,
        "fullAddress": "557南投縣竹山鎮大智路52號",
        "continents": "本會",
        "latitude": 23.7602475,
        "longitude": 120.6940623
    },
    {
        "affiliateId": "8a818cf519b813f80119b814179f00c3",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b814179f00c3",
        "affiliateName": "嘉義聯絡處",
        "affiliateEngName": "",
        "buildingName": "嘉義靜思堂",
        "affiliateAttribute": "聯絡處",
        "postalCodeName": "嘉義市",
        "address": "彌陀路26號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "是",
        "phone1": "05-2777852",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市彌陀路26號",
        "continents": "本會",
        "latitude": 23.4653463,
        "longitude": 120.4689758
    },
    {
        "affiliateId": "2c90819488e5e523018975e4350800e1",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b814179f00c3",
        "affiliateName": "嘉義志業園區",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "尚未成立",
        "postalCodeName": "嘉義市",
        "address": "嘉義市博愛路2段469巷19號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "05-2353318",
        "isStop": "未啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市嘉義市博愛路2段469巷19號",
        "continents": "本會",
        "latitude": 23.4740924,
        "longitude": 120.4328501
    },
    {
        "affiliateId": "8a81958b312f09b1013130f188f20018",
        "nation": "臺灣",
        "mission": "靜思人文",
        "parentAffiliate": "2D31312E30302E35382E353430343235",
        "affiliateName": "嘉義靜思書軒",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "靜思書軒",
        "postalCodeName": "嘉義市",
        "address": "彌陀路28號",
        "canPublishExternal": "是",
        "canPublishInternal": "是",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2012/12/18",
        "modifyOpId": "蔡奇珊",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市彌陀路28號",
        "continents": "本會",
        "latitude": 23.4654729,
        "longitude": 120.4682491
    },
    {
        "affiliateId": "2D30382E35352E30352E353039303637",
        "nation": "臺灣",
        "mission": "醫療",
        "parentAffiliate": "2D31342E35302E35362E393830343830",
        "affiliateName": "嘉義慈濟居家式長照機構",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "長照機構",
        "postalCodeName": "嘉義市",
        "address": "延平街448號三樓之4",
        "canPublishExternal": "否",
        "canPublishInternal": "是",
        "canEngPublishExternal": "是",
        "canEngPublishInternal": "是",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2023/09/07",
        "modifyOpId": "atc00099",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市延平街448號三樓之4",
        "continents": "本會",
        "latitude": 23.4761399,
        "longitude": 120.4416689
    },
    {
        "affiliateId": "8a81958b46a3642e0146a73dafb50001",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b814179f00c3",
        "affiliateName": "國華環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "嘉義市",
        "address": "嘉義市西區撫順三街119號 ",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2014/06/17",
        "modifyOpId": "張涵鈞",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市嘉義市西區撫順三街119號 ",
        "continents": "本會",
        "latitude": 23.4910831,
        "longitude": 120.4414567
    },
    {
        "affiliateId": "8a8195986dcf9275017176f56e610009",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b814179f00c3",
        "affiliateName": "嘉義志業園區環保教育站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保教育站",
        "postalCodeName": "嘉義市",
        "address": "嘉義市博愛路2段469巷19號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "2020/04/14",
        "modifyOpId": "張涵鈞",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市嘉義市博愛路2段469巷19號",
        "continents": "本會",
        "latitude": 23.4740924,
        "longitude": 120.4328501
    },
    {
        "affiliateId": "2D31312E34362E30322E363439383733",
        "nation": "臺灣",
        "mission": "慈善",
        "parentAffiliate": "8a818cf519b813f80119b814179f00c3",
        "affiliateName": "福德環保站",
        "affiliateEngName": "",
        "buildingName": "",
        "affiliateAttribute": "環保站",
        "postalCodeName": "嘉義市",
        "address": "世賢路二段196巷98號",
        "canPublishExternal": "否",
        "canPublishInternal": "否",
        "canEngPublishExternal": "否",
        "canEngPublishInternal": "否",
        "canPublishLiteratureAndHistory": "否",
        "phone1": "",
        "isStop": "啟用",
        "stopDate": "",
        "stopOpId": "",
        "modifyDate": "",
        "modifyOpId": "",
        "province": "嘉義市",
        "postalCode": 600,
        "fullAddress": "600嘉義市世賢路二段196巷98號",
        "continents": "本會",
        "latitude": 23.480001,
        "longitude": 120.4166664
    },
]

updated_data = update_json_with_place_details(api_key, json_data)
print(json.dumps(updated_data, ensure_ascii=False, indent=4))
```
