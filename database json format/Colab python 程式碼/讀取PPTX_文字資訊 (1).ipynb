```python
# Specify the path to your PowerPoint file on Google Drive
pptx_path = '/content/drive/MyDrive/but but but/keep/activity table/泰國(1).pptx'
json_path = '/content/drive/MyDrive/but but but/keep/coba.json'

# Load the presentation
presentation = Presentation(pptx_path)

# Function to extract text from each slide
def extract_text_from_slide(slide):
    text = []
    for shape in slide.shapes:
        if hasattr(shape, "text"):
            text.append(shape.text)
    return text

# Extract text from all slides and format into JSON array
new_text_json = []
for slide in presentation.slides:
    slide_text = extract_text_from_slide(slide)
    slide_json = {
        "country": slide_text[7] if len(slide_text) > 7 else "",
        "title": slide_text[6] if len(slide_text) > 6 else "",
        "description": slide_text[8] if len(slide_text) > 8 else "",
        "date": slide_text[10] if len(slide_text) > 10 else "",
        "location": slide_text[9] if len(slide_text) > 9 else "",
        "photographer": slide_text[11] if len(slide_text) > 11 else ""
    }
    new_text_json.append(slide_json)

# Read existing JSON data if the file exists
if os.path.exists(json_path):
    with open(json_path, 'r', encoding='utf-8') as f:
        existing_data = json.load(f)
else:
    existing_data = []

# Append the new data
existing_data.extend(new_text_json)

# Save the combined JSON output to the file
with open(json_path, 'w', encoding='utf-8') as f:
    json.dump(existing_data, f, ensure_ascii=False, indent=4)

print(f"Extracted text has been saved to {json_path}")
```

```python
!pip install python-pptx

from pptx import Presentation
import json

# Define the function to extract text information from a PPTX file
def extract_pptx_info(pptx_path, countries):
    presentation = Presentation(pptx_path)
    slides_data = []

    for slide_index, slide in enumerate(presentation.slides):
        slide_info = {}
        if slide_index < len(countries):
            slide_info["country"] = countries[slide_index]
        else:
            slide_info["country"] = f"Country {slide_index + 1}"

        text_boxes = {shape_index: shape.text_frame.text for shape_index, shape in enumerate(slide.shapes) if shape.has_text_frame}

        slide_info["eng-title"] = text_boxes.get(10, "").strip()
        slide_info["zh-title"] = text_boxes.get(9, "").strip()
        slide_info["description"] = text_boxes.get(8, "").strip()

        if slide_info["eng-title"] and slide_info["zh-title"] and slide_info["description"]:
            slides_data.append(slide_info)

    return slides_data

# Define the list of countries
countries = [
    "Australia",
    "Papua New Guinea",
    "Thailand",
    "Laos",
    "Turkey",
    "Turkey",
    "Iraq",
    "Iran",
    "Afghanistan",
    "Azerbaijan",
    "Syria",
    "Chechnya",
    "Mongolia",
    "India",
    "Nepal",
    "Bangladesh",
    "Pakistan",
    "Ukraine",
    "Portugal",
    "France",
    "Kosovo",
    "Morocco",
    "Sierra Leone",
    "Rwanda",
    "Congo",
    "Ethiopia",
    "St. Vincent",
    "Puerto Rico",
    "Nicaragua",
    "Peru",
    "Colombia"
]


# Path to the PPTX file
pptx_path = "/content/drive/MyDrive/but but but/keep/title.pptx"

# Extract the information
slides_data = extract_pptx_info(pptx_path, countries)

# Convert the data to JSON format
slides_json = json.dumps(slides_data, ensure_ascii=False, indent=2)

# Save the JSON to a file
with open("/content/slides_data.json", "w", encoding="utf-8") as json_file:
    json_file.write(slides_json)

# Print the JSON
print(slides_json)
```
